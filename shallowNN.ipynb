```python
import numpy as np
import matplotlib.pyplot as ply
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def softmax(x):
    a=np.exp(x)
    return a/np.sum(a)
```

```python
class ShallowNN:
    def __init__(self, num_input, num_hidden, num_output):
        self.W_h=np.zeros((num_hidden, num_input),dtype=np.float32)
        self.b_h=np.zeros((num_hidden,),dtype=np.float32)
        self.W_o=np.zeros((num_output,num_hidden),dtype=np.float32)
        self.b_o=np.zeros((num_output),dtype=np.float32)
    def __call__(self,x):
        h=sigmoid(np.matmul(self.W_h,x)+self.b_h)
        return softmax(np.matmul(self.W_o,h)+self.b_o)
        
```

```python
inputs=np.load("ch2_dataset/inputs.npy")
```

```python
label=np.load("ch2_dataset/labels.npy")
```

```python
model=ShallowNN(2,128,10)
```

```python
model.W_h=np.load("ch2_parameters/W_h.npy")
model.b_h=np.load("ch2_parameters/b_h.npy")
model.W_o=np.load("ch2_parameters/W_o.npy")
model.b_o=np.load("ch2_parameters/b_o.npy")
```

```python
outputs=list()
for pt, lab in zip(inputs,label):
    output=model(pt)
    outputs.append(np.argmax(output))
    print(np.argmax(output),lab)
    
```

```python
ply.figure()
for idx in range(10):
    mask=label==idx
    ply.scatter(inputs[mask,0],inputs[mask,1])
```

```python
ply.figure()
for idx in range(10):
    mask=outputs==idx
    ply.scatter(inputs[mask,0],inputs[mask,1])
```
